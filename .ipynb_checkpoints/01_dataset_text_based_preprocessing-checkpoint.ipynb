```python
import pandas as pd
import re
from utils.ml import *
from utils.alloys import order_df, get_elements_and_compositions
```

```python
df = pd.read_excel("d_max_dataset.xlsx")
display(df)
```

```python
df = df[df["bmg_alloy"].notna()]
df = df[df["actual_d_max"].notna()]
df = df[df["paper_sno"].notna()]
df["paper_sno"] = pd.to_numeric(df["paper_sno"], downcast="integer")
```

```python
display(df)
```

```python
df = df[df.paper_sno != 4]
df = df[df.paper_sno != 8]
df["actual_d_max"] = pd.to_numeric(df["actual_d_max"], errors="coerce")
df = df[df["actual_d_max"].notna()]
df = df.reset_index(drop=True)
display(df)
```

```python
def ends_with_square_bracket_digit_square_bracket(input_string):
    pattern = r'.*\[\d+\]$'
    match = re.match(pattern, input_string)
    return bool(match)

def trim_pattern_from_end(input_string):
    pattern = r'\[\d+\]$'
    trimmed_string = re.sub(pattern, '', input_string)
    return trimmed_string.strip()

for index, row in df.iterrows():
    try:
        alloy = row["bmg_alloy"]
        if ends_with_square_bracket_digit_square_bracket(alloy):
            print(f"{alloy}")
            trimmed_alloy = trim_pattern_from_end(alloy)
            df.at[index, "bmg_alloy"] = trimmed_alloy
    except Exception as e:
        print(e)
```

```python
def contains_special_char(input_string):
    pattern = r'[\(\)\[\]{}\\]'
    match = re.search(pattern, input_string)
    return bool(match)

for index, row in df.iterrows():
    try:
        alloy = row["bmg_alloy"]
        if contains_special_char(alloy):
            print(alloy)
            df.drop(index, inplace=True)
    except Exception as e:
        print(e)
```

```python
df = df.reset_index(drop=True)
display(df)
```

```python
df = order_df(df)
display(df)
```

```python
for index, row in df.iterrows():
    try:
        alloy = row["bmg_alloy"]
        elements, compositions = get_elements_and_compositions(alloy)
        if sum(compositions) < 0.90:
            print(alloy)
            df.drop(index, inplace=True)
    except Exception as e:
        print(e)
```

```python
group_alloys_by_paper = {}
for index, row in df.iterrows():
    try:
        alloy = row["bmg_alloy"]
        paper_sno = row["paper_sno"]
        d_max = row["actual_d_max"]
        if alloy in group_alloys_by_paper:
            group_alloys_by_paper[alloy]["paper_sno"].add(paper_sno)
            group_alloys_by_paper[alloy]["actual_d_max"].add(d_max)
        else:
            group_alloys_by_paper[alloy] = {}
            group_alloys_by_paper[alloy]["paper_sno"] = {paper_sno}
            group_alloys_by_paper[alloy]["actual_d_max"] = {d_max}
    except Exception as e:
        print(e)
```

```python
rows = []

# Iterate over the dictionary and create rows for the DataFrame
for bmg_alloy, values in group_alloys_by_paper.items():
    row = [bmg_alloy, list(values['paper_sno']), list(values["actual_d_max"])]
    rows.append(row)

# Create the DataFrame with appropriate column names
df_unique = pd.DataFrame(rows, columns=['bmg_alloy', 'paper_sno', "actual_d_max"])

display(df_unique)
```

```python
conflicting_dmax_df = df_unique[df_unique["actual_d_max"].apply(len) > 1]
display(conflicting_dmax_df)
```
