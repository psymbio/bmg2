```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

element_symbols = df.columns

# Convert the DataFrame to a numpy array
element_vectors_array = df.to_numpy()

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_vectors = pca.fit_transform(element_vectors_array)

# Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(reduced_vectors[:, 0], reduced_vectors[:, 1], marker='o')

for i, symbol in enumerate(element_symbols):
    plt.text(reduced_vectors[i, 0], reduced_vectors[i, 1], symbol, fontsize=20)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Latent Space Visualization of Element Distances')
plt.show()
```

```python
element_vectors_array = df.to_numpy()

# Apply PCA
pca = PCA()
pca.fit(element_vectors_array)

# Get the explained variance ratio for each principal component
explained_variance_ratio = pca.explained_variance_ratio_

# Plot explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(explained_variance_ratio))
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.grid(True)
plt.show()

# Get the first principal component
first_principal_component = pca.components_[0]

# Display the importance of each element in the first principal component
component_importance = pd.Series(first_principal_component, index=df.columns)
component_importance = component_importance.abs().sort_values(ascending=False)
print(component_importance)
```
